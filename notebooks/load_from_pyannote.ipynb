```python
from pyannote.audio import Model

model = Model.from_pretrained("pyannote/segmentation-3.0")
```

```python
from collections import OrderedDict
state_dict = model.state_dict()
items = list(state_dict.items())
sincnet_items = []
for item in items:
    if item[0].startswith('sincnet.'):
        sincnet_items.append(
            (item[0][8:], item[1])
        )


len(sincnet_items)
```

```python
sincnet_items[5]
```

```python
sincnet_state_dict = OrderedDict()

sincnet_state_dict['wav_norm1d.weight'] =  sincnet_items[0][1]
sincnet_state_dict['wav_norm1d.bias'] =  sincnet_items[1][1]
sincnet_state_dict['conv1d.0._low_hz'] =  sincnet_items[2][1]
sincnet_state_dict['conv1d.0._band_hz'] =  sincnet_items[3][1]
sincnet_state_dict['conv1d.0._window'] =  sincnet_items[4][1]
sincnet_state_dict['conv1d.0._n'] =  sincnet_items[5][1].reshape(1, -1)

# copy the rest of the parameters
for i in range(6, len(sincnet_items)):
    sincnet_state_dict[sincnet_items[i][0]] = sincnet_items[i][1]

sincnet.load_state_dict(sincnet_state_dict)
```

```python
import torch
torch.save(sincnet.state_dict(), '../models/sincnet.pth')
```
